# 소트하는 세가지 방법

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(123456789)
```

# sort method를 이용한 소트

sort method는 inplace function이다.

```python
x = np.random.uniform(0.,1.,5)
x.sort() # inplace function
x
```

sort method는 inplace function이다.
소트한 것을 어싸인하면 않된다.

```python
x = np.random.uniform(0.,1.,5)
x = x.sort() # inplace function
x
```

sort method는 inplace function이다.
소트한 것을 어싸인하면 않된다.
sort method는 inplace function임을 나타내기 위하여, None을 리턴하기 때문이다.

```python
x = np.random.uniform(0.,1.,5)
x = x.sort() # inplace function
print(x)
```

```python
x = np.random.uniform(-1.,1.,100).sort()
print(x)
```

# np.argsort를 이용한 소트

```python
x = np.random.uniform(0.,1.,5)
id_x = np.argsort(x)

print(x)
print(id_x)
print(x[id_x])
print(x[np.argsort(x)])
```

# np.sort를 이용한 소트

```python
x = np.random.uniform(0.,1.,5)
x_sorted = np.sort(x)

print(x)
print(x_sorted)
```

# $y=x^2$ 그래프 그리기

```python
x = np.random.uniform(-1.,1.,100)
y = x**2
plt.plot(x,y)
```

```python
x = np.random.uniform(-1.,1.,100)
y = x**2
plt.plot(x,y,'.')
```

```python
x = np.random.uniform(-1.,1.,100)
x.sort()
y = x**2
plt.plot(x,y)
```

```python
x = np.sort( np.random.uniform(-1.,1.,100) )
y = x**2
plt.plot(x,y)
```
